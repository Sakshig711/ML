```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
```

```python
df = pd.read_csv('cities_r2.csv')
```

```python
df.head()
```

```python
x = df.iloc[:,[7,19]]
x
```

```python
plt.title('Unclustered Data')
plt.xlabel('Total graduates')
plt.ylabel('Total Population')
plt.scatter(x['total_graduates'],x['0-6_population_total'])
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 5)
km.fit_predict(x)
```

```python
km.inertia_
```

```python
sse = []
for k in range(1,16):
    km = KMeans(n_clusters = k)
    km.fit_predict(x)
    sse.append(km.inertia_)
warnings.filterwarnings('ignore')
```

```python
sse
```

```python
#plot elbow method
plt.title('Elbow Method')
plt.xlabel('Value of K')
plt.ylabel('SSE')
plt.grid()
plt.xticks(range(1,16))
plt.plot(range(1,16), sse, marker = '.', color = 'red')
```

```python
from sklearn.metrics import silhoutte_score
```

```python
from sklearn.metrics import silhouette_score
```

```python
silh = []
for k in range(2,16):
    km = KMeans(n_clusters = k)
    labels = km.fit_predict(x)
    score = silhouette_score(x,labels)
    silh.append(score)
```

```python
silh
```

```python
plt.title('Silhouette Method')
plt.xlabel('Value of K')
plt.ylabel('Silhoutte Score')
plt.grid()
plt.xticks(range(2, 16))
plt.bar(range(2, 16), silh, color = 'pink')
#below graph shows k=5 is best clustering
```

```python
km = KMeans(n_clusters = 5, random_state = 0)
labels = km.fit_predict(x)
labels
```

```python
#Seeing clustered and unclustered same?
plt.figure(figsize = (16,9))

plt.subplot(1,2,1)
plt.title('Unclustered Data')
plt.xlabel('Total graduates')
plt.ylabel('Toatl Population')
plt.scatter(x['total_graduates'], x['0-6_population_total'])

plt.subplot(1,2,2)
plt.title('Clustered Data')
plt.xlabel('Total graduates')
plt.ylabel('Toatl Population')
plt.scatter(x['total_graduates'], x['0-6_population_total'], c=labels)
```
